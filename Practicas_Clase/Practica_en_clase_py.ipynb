```python
#Crear una clase llamada Persona
class Persona:
  #Crear un metodo llamado constructor para inicializar mis atributos
  def __init__(self,nombre,edad):
    #Crear los atributos de la clase Persona
    self.nombre=nombre
    self.edad=edad
  #Crear un metodo para presentar informacion
  def presentar(self):
    print(f"Hola, me llamo {self.nombre} y tengo {self.edad} años")
#Ingreso de informacion al usuario
nombre=input("Ingrese su nombre: ")
edad=int(input("Ingrese su edad: "))
#Crear un objeto (persona1) a partir de una clase e instanciarla (Persona)
persona1=Persona(nombre,edad)
persona1.presentar()

```
